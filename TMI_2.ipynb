**For state rho = p/3|100+010+001><100+010+001| + q |111><111| + (1-p-q)|000><000|**

```python
zero = [[1],
       [0]]
one = [[0],
      [1]]
state_0 = np.array(zero)
state_1 = np.array(one)
```

```python
state_00 = np.kron(state_0,state_0)
state_000 =np.kron(state_0,state_00)
state_11 = np.kron(state_1,state_1)
state_111=np.kron(state_1,state_11)
state_100 = np.kron(state_1,state_00)
state_001 = np.kron(state_00,state_1)
state_01 = np.kron(state_0,state_1)
state_010 = np.kron(state_01,state_0)
```

```python
p = []
import random 
for i in range(100) :
    i = random.uniform(0,1)
    p.append(i)
p.append(0)
p.append(1)
p
```

```python
q1 =[]
for i in range (len(p)):
    i =random.uniform(0,1)
    q1.append(i)
for i in range(len(p)):
    if (p[i] + q1[i]) > 1:
        q1[i] = 1-p[i]
    else:
        q1[i] = q1[i]
    q1.append(q1[i])
    for j in range(len(p)):
        if q1[j] < 0:
            q1[j] = abs(q1[j])
q = q1[:100]
q.append(0)
q.append(1)
q
```

```python
list_Rho123=[]
for i in range(len(q)):
    Rho_123 = (p[i]/3)*np.outer(state_100+state_010+state_001,np.conj(state_100+state_010+state_001))+q[i]*np.outer(state_111,np.conj(state_111))+(1-p[i]-q[i])*np.outer(state_000,np.conj(state_000))
    list_Rho123.append(Rho_123)
list_Rho123
```

```python
def extract_elements_4(Rho_123,indices):
    # Create a new 4x4 matrix filled with zeros
    four_by_four = [[ 0 for _ in range(4)] for _ in range(4)]
    # Iterate over the elements of the 4x4 matrix
    for i in range (4):
        for j in range (4):
            row_idx, col_idx = indices[i * 4 + j]
            four_by_four[i][j] = Rho_123[row_idx][col_idx]
    return four_by_four
```

```python
def extract_elements_2(matrix,indices):
    two_by_two =[[0 for _ in range(2)] for _ in range(2)]
    for i in range(2):
        for j in range(2):
            row_idx, col_idx = indices [i*2 +j]
            two_by_two[i][j]= matrix[row_idx][col_idx]
    return two_by_two
```

**finding von-neumann entropy**

```python
def Entropy(rdm):
    lmbda = np.round(np.linalg.eigvals(rdm),4)
    if  lmbda.any() <  0:
        lmbda == 0
    S = 0
    for i in range (len(lmbda)):
        if lmbda[i] == 0:
            S = S + 0
        else:
            S = np.round(S - lmbda[i]*np.log2(lmbda[i]),4)
        if S < 0 :
            S==0
    return S
```
